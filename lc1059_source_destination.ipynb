1059. All Paths from Source Lead to Destination

https://leetcode.com/problems/all-paths-from-source-lead-to-destination/


```python
class Solution:
    def leadsToDestination(self, n: int, edges: List[List[int]], start: int, dest: int) -> bool:
        graph = [set() for _ in range(n)]
        for u, v in edges:
            graph[u].add(v)
        WHITE = 0
        GREY = 1
        BLACK = 2
        visit = [WHITE] * n
        def dfs(s):
            visit[s] = GREY 
            if s != dest and len(graph[s]) == 0:
                return False 
            for v in graph[s]:
                # detect back edges
                if visit[v] == GREY:
                    return False
                # if not visited, then visit
                if visit[v] == WHITE and dfs(v) is False:
                    return False 
            visit[s] = BLACK
            return True
            
        return dfs(start)
# Time: O(V+E)
# Space: O(V+E)
```

```python
n = 4; edges = [[0,1],[0,2],[1,3],[2,3]]; source = 0; destination = 3
Solution().leadsToDestination(n,edges,source,destination)
```
