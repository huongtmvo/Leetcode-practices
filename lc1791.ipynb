1791. Find Center of Star Graph

https://leetcode.com/problems/find-center-of-star-graph/

```python
# Solution 1: 
# Reasoning: The node at the center of the graph has a length of adjacency 
# node lists equal to the length of the edges
from typing import List
class Solution:
    def findCenter(self, edges: List[List[int]]) -> int:
        # number of nodes n
        n = len(edges) + 1
        adj = [[] for _ in range(n + 1)]

        # Construct the adjacency list
        for i in range(n-1):
            adj[edges[i][0]].append(edges[i][1])
            adj[edges[i][1]].append(edges[i][0])

        # If the length of the adjacency list = number of edges-> center of the star graph
        for v in range(1, n+1):
            if len(adj[v]) == n -1:
                return v
        return -1
```
