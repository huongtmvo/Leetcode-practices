841. Keys and Rooms

https://leetcode.com/problems/keys-and-rooms/

```python
from typing import List
class Solution:
    def canVisitAllRooms(self, rooms: List[List[int]]) -> bool:
        n = len(rooms)
        visited = set()
        def dfs(r):
            visited.add(r)
            for u in rooms[r]:
                if u not in visited:
                    dfs(u)
        
        dfs(0)
        return True if len(visited) == n else False
```

```python
from typing import List
class Solution2:
    def canVisitAllRooms(self, rooms: List[List[int]]) -> bool:
            n = len(rooms)
            self.visited = set()
            self.dfs(0, rooms)
            return True if len(self.visited) == n else False

    def dfs(self, r, rooms):
            self.visited.add(r)
            for u in rooms[r]:
                if u not in self.visited:
                    self.dfs(u, rooms)
```

```python
rooms = [[1],[2],[3],[]]
Solution2().canVisitAllRooms(rooms)
```
