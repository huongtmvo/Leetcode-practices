252. Meeting Rooms

https://leetcode.com/problems/meeting-rooms/

```python
from collections import deque
from typing import List
class Solution:
    def canAttendMeetings(self, intervals: List[List[int]]) -> bool:
        intervals.sort(key=lambda x: x[0])
        n = len(intervals)
        if n <= 1: return True
        queue = deque([intervals[0]])
        for i in range(1,n):
            last = queue.popleft()
            if intervals[i][0] < last[1]:
                return False
            queue.append(intervals[i])
        return True
# Time complexity: O(N)
# Space complexity: O(1) (there is only 1 node at a queue at one time)
```

```python
intervals = [[0,30],[5,10],[15,20]]
Solution().canAttendMeetings(intervals)
```

```python
intervals = [[7,10],[2,4]]
Solution().canAttendMeetings(intervals)
```
