https://leetcode.com/problems/intersection-of-two-arrays/

```python
class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        set1 = set(nums1)
        res = set()
        for num in nums2:
            if num in set1:
                res.add(num)
        return res
        
    def intersection(self, nums1, nums2):
        return list(set(nums1) & set(nums2))
```
